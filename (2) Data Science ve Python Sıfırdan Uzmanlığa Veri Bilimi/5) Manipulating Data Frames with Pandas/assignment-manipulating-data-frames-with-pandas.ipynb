## Index kolonunu değiştirme
İlk yapacağımız işlen 'Unnamed: 0' kolonumuzun ismini değiştirmek olacak.

```python
data = data.rename(columns = {'Unnamed: 0' : 'id'})
# Değişip değişmediğini kontrol edelim
data.head()
```

Elimizdeki **id** isimli kolonu **index** olarak kullanmak için: 

```python
# id kolonundaki sayılar düzensiz şekilde ilerlediği için  başka bir dataframe üzerinde örnek olarak index değişikliği yaptık.
data_ornek = data.set_index('id')
# Kontrol
data_ornek.head()
```

Index olarak kullandığımız kolonun adını değiştirelim.

```python
data.index.name = 'index_name'
```

## Slicing Data Frames
Tanımladığımız birinci kolon ile ikinci kolon arasındaki ilk 5 veriyi listeleyelim.

```python
data.loc[1:5,'course_title':'course_rating']
```

İlk 5 veriyi tersten listeleyelim.

```python
data.loc[5:1:-1,'course_title':'course_rating']
```

Birinci kolondan başlayıp geri kalan bütün kolonları listeleyelim.

```python
data.loc[1:10,'course_rating':]
```

## Filtering Data Frames
Data içerisinde oluşturacağımız filtreye uyan veriyi listeyelim.

```python
data_boolean = data.course_rating > 4.5
data[data_boolean]
```

Birden fazla filtre oluşturup data üzerinde kullanalım.

```python
first_filter = data.course_rating>4.5
second_filter = data.course_Certificate_type == 'SPECIALIZATION'
third_filter = data.course_difficulty == 'Beginner'
data[first_filter & second_filter & third_filter].head(10)
```

Peki filter sonucu gelen veride sadece belirli bir kolonu çekmek istediğimizde ? 

```python
data[first_filter & second_filter & third_filter].course_title.head(10)
```

## Hierarchical Indexing

Burada ise data içerisindeki kolonları kullanarak indexleme işlemini nasıl yapacağımıza bakalım. 

```python
data_hierarchical_indexing = data.set_index(['course_rating','course_students_enrolled'])
data_hierarchical_indexing.head(10)
```

Çıkan sonuca baktığımızda gördüğümüz gibi önce veriyi 1. index değeri '**course_rating**' e göre gruplandırdı ardından 2. index değeri olan '**course_students_enrolled**' a göre de bir iç gruplandırma yaptı.
